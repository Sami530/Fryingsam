```python
#Importing the matplotlib library
import matplotlib.pyplot as plt
#import the numpy library
import numpy as np
```

```python
#lets get some randome values with the help of numpy
x = np.linspace(0,10,100) # will get the 100 random evenly spaced values from 0-10
y = np.sin(x)
z = np.cos(x)
```

```python
print(x)
```

```python
print(y)
```

```python
#plotting the sin wave
plt.plot(x,y)
plt.show()
```

```python
#plotting the cosine wave
plt.plot(x,z)
plt.show()
```

```python
#adding the titles to the label x and y axis
plt.plot(x,y)
plt.xlabel("angle")
plt.ylabel("sine value")
plt.title("The sine Wave")
plt.show()
```

```python
#adding the title and x ,y axis
plt.plot(x,z)
plt.xlabel("angle")
plt.ylabel("cosine value")
plt.title("The cosine wave")
plt.show()
```

```python
#parabola
x = np.linspace(-10,10,20)
y = x**2
plt.plot(y)
```

```python
plt.plot(x,y,"r+")
plt.show()
```

```python
plt.plot(x,y,"g.")
plt.show()
```

```python
#plotting multiple lines on single view
x = np.linspace(-5,5,50)
plt.plot(x,np.sin(x),"g-")
plt.plot(x,np.cos(x),"r.")
plt.show()
```

```python
#Bar Plot
fig = plt.figure()
ax = fig.add_axes([0,0,1,1]) #here 0,0 are origin and 1,1 represent the height and width of the rect
languages = ["English","French","Spanish","Latin","German"]
people = [100,50,150,40,70]
ax.bar(languages,people)
plt.xlabel("LNG")
plt.ylabel("No of people")
plt.show()
```

Pie Chart

```python
fig1 =plt.figure()
ax = fig1.add_axes([0,0,1,1])
languages = ["English","French","Spanish","Latin","German"]
people = [100,50,150,40,70]
ax.pie(people,labels=languages,autopct="%1.1f%%") 
plt.show()
```

Scatter-Plot

```python
x = np.linspace(0,10,30)#used for clustering
y = np.sin(x)
z = np.cos(x)
fig2 = plt.figure()
ax = fig2.add_axes([0,0,1,1])
ax.scatter(x,y,color="g")
ax.scatter(x,z,color="r")
plt.show()
```

3-D Scatter plot

```python
fig3 = plt.figure()
ax = plt.axes(projection="3d")
z = 20 * np.random.random(100) #100random val
x = np.sin(z)
y = np.cos(z)
ax.scatter(x,y,z,c=z,cmap="Blues")
plt.show()
```

Scatter plot

```python
x=[1,2,3,4,5]
y=[2,4,6,8,10]
plt.title("Scatter plot")
plt.scatter(x,y,color='Red',Label="Scatter Plot",s=150,marker='*' )
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.legend(loc=2)
plt.show()
```

Pie Chart Once again with explode parameters

```python
slices = [30,40,60,80,90]
activities = ["sleeping","studying","Eating","Playing","Gaming"]
cols = ["Red","Green","Purple","Blue","Yellow"]

plt.pie(slices,labels=activities,colors=cols,autopct="%1.2f%%",explode=(0.2,0.2,0,0,0))

plt.show()
```

Subplots

```python
fig = plt.figure()
names = ['A','B','C']
values = [19,20,43]
values_2 = [45,67,87]

ax = fig.add_subplot(121)
ax2 = fig.add_subplot(122)
ax.bar(names,values)
ax2.bar(names,values_2)

plt.show()
```

```python
fig=plt.figure()
names=['A','B','C']
values=[19,50,29]
values_2=[48,19,41]

ax=fig.add_subplot(211)## first number=how many rows,second number=how many columns,third number=number of plots the u want to add

ax2=fig.add_subplot(212)

ax.bar(names,values)

ax2.bar(names,values_2)
plt.show()
```
